```python
df_confirm = df[df['Статус Выкупа'] != '--']
```

```python
px.histogram(df_confirm,
             x = 'Оператор',
             color = 'Категория',
             barmode = 'group',
             text_auto = True,
             title = 'Количество подтвержденных заказов'
             )
```

```python
paid_df = df[df['Статус Выкупа'] == 'Выкуплен']
unpaid_df = df[df['Статус Выкупа'] == '--']

paid = paid_df.groupby(['Оператор', 'Категория']).agg({'Статус подтверждения' : 'count'}).reset_index()
unpaid = unpaid_df.groupby(['Оператор', 'Категория']).agg({'Статус подтверждения' : 'count'}).reset_index()
new_df = df.groupby(['Оператор', 'Категория']).agg({'Статус подтверждения' : 'count'}).reset_index()

paid.columns = ['Оператор1', 'Категория1', 'Выкуплено']
unpaid.columns = ['Оператор2', 'Категория2', 'Отменено']

new_df = pd.concat([new_df, paid], axis=1)
new_df = pd.concat([new_df, unpaid], axis=1)
new_df = new_df.drop(['Оператор1', 'Категория1', 'Оператор2', 'Категория2'], axis= 1)

difference = new_df

difference['Процент выкупа'] = (round(difference['Выкуплено'] / difference['Статус подтверждения'] * 100, 2))
difference['Процент отмен'] = (round(difference['Отменено'] / difference['Статус подтверждения'] * 100, 2))
difference = difference.drop(['Статус подтверждения', 'Выкуплено', 'Отменено'], axis= 1)
difference = difference.sort_values('Процент выкупа', ascending=False).reset_index(drop = True)
```

```python
display(difference)
```

```python
px.histogram(difference,
             x = 'Оператор',
             y = 'Процент выкупа',
             color = 'Категория',
             barmode = 'group',
             text_auto = True,
             title = 'Процент выкупа'
             )
```

```python
px.histogram(difference,
             x = 'Оператор',
             y = 'Процент отмен',
             color = 'Категория',
             barmode = 'group',
             text_auto = True,
             title = 'Процент отмен'
             )
```
